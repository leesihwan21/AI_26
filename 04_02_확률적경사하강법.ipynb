```python
# 확률적 경사 하강법 (실제 산에서 내려오는 것 처럼 가파른 길을 찾아서 천천히 조금씩 내려옴)
# 확률적 : 무작위, 랜덤(훈련 세트에서 랜덤하게 하나의 샘플을 추출)
# 경사 : 기울기
# 하강 : 내려가는
# 확률적 경사 하강법은 훈련 세트에서 랜덤하게 하나의 샘플을 선택해서 가파른 경사를 조금 내려감
# 위 방법을 반복수행함.
# 만약에 모든 샘플을 다 사용해도 다 내려오지 못 할 경우
#   처음부터 다시 시작함
#   훈련세트에 모든 샘플을 다시 채워 넣음
#   다시 랜덤하게 하나의 샘플을 선택하여 이어서 경사를 내려감
# 만족할만한 위치에 도달 할 때까지 계속 내려감
# 훈련세트를 한 번에 모든 데이터를 사용하는 과정을 에포크(epoch)라고 함.
#   일반적인 경사하강법은 수십, 수백번 이상 에포크를 수행함
# 만약, 실제로 등산이라고 가정하고 산에서 내려온다고 생각했을 때 집으로 돌아가려면
# 등산로의 입구까지 내려가야 한다.
# 만약 어떤 산길도 척척 내려갈 수 있는 초능력이 있다면 가장 빠른 길을 선택하는 것이 최선임
# 가장 빠른 길은 경사가 가장 가파른 길이다!!!!!
```

```python
# 실제로 산에서 내려올 때는 천천히 조금씩 내려와야 한다. -> 경사 하강법
# 가장 가파른 길을 찾아 내려오지만 조금씩 내려오는 것이 중요함 -> 훈련법
# 훈련 세트를 사용해서 모델을 훈련하기 때문에 경사 하강법도 당연히 가장 가파른 길을 찾을 것!
# 그런데 전체 샘플을 사용하지 않고 딱 하나의 샘플을 훈련 세트에서 랜덤하게 골라 가장 가파른 길을 찾음
# -> 확률적(무작위, 랜덤)
# 결론 : 훈련세트에서 랜덤하게 하나의 샘플을 선택해서 가파른 경사를 조금씩 내려간 다음 훈련세트에서
# 랜덤하게 또 다른 샘플을 하나 선택해서 경사를 조금 내려감 -> 전체 샘플을 모두 사용할 때까지 진행
```

```python
# 위 방법대로 샘플 1개를 이용하면 너무 무책임하게 내려갈 수 있다.
# 잘못된 훈련으로 결과가 나오면 돌아갈 수 없다.
# 이러한 문제점을 해소하려고 무작위로 몇개의 샘플을 선택해서 경사를 내려간다면
  # 미니 배치 경사하강법 : 여러개의 샘플을 사용해서 경사하강진행
  # 배치 경사하강법 : 극단적으로 한 번 경사를 따라 이동 (전체 샘플사용)
    # 단점 : 컴퓨터 자원을 너무 많이 사용한다.
    # 한 번에 전체 데이터를 모두 읽을 수 없다.
# 확률적 경사 하강법은 훈련 세트를 사용해 산 아래에 있는 최적의 장소로 조금씩 이동
# 이 때문에 훈련 데이터가 모두 준비되어 있지 않고, 매일매일 업데이트(최신) 되어도 학습을 이어나갈 수 있다.
# 다시 산꼭대기 부터 시작할 필요가 없다. -> 신경망 알고리즘을 사용하면 된다.

# 손실 함수 (Loss function) -> 머신러닝 알고리즘이 얼마나 엉터리인지 측정하는 기준 (작을수록 좋음)
#                           -> 하지만 어떤 값이 최소값인지 알지 못함.
#                           -> 가능한 많이 찾아보고 만족할 만한 수준이면 인정해야 함.
#                           -> 이 값을 찾아서 조금씩 이동

# 손실 함수 공식 : Loss = -log(p) -> p = 모델이 "1"이라고 예측한 확률
# 예를 들면 정답이 1일 때 : Loss = -log(p) (확률이 1에 가까워질수록 손실은 0에 수렴한다는 뜻)
# ->        정답이 0일 때 : Loss = -log(p) (확률이 0에 가까워질수록 손실은 0에 수렴한다는 뜻)

# 비용 함수 (Cost function) -> 손실함수의 다른 말. (손실함수는 샘플 하나에 대해서 손실을 정의)
#                           -> 비용함수는 훈련세트에 있는 모든 샘플에 대한 손실함수의 합

# 분류에서 손실은 정답을 못 맞히는 것

# 에포크(epoch : 시대) -> 훈련세트를 한 번 모두 사용하는 과정(전체 데이터)
```

```python
# 정답이 1인데
# 모델이 높은 확률(0.9, 0.8) 로 맞추면 -> 손실이 작다
# 모델이 낮은 확률(0.3, 0.2) 로 맞추면 -> 손실이 크다

# 로그값은 0~1 사이에서 항상 음수이기 때문에
# log(0.9) ≈ -0.1
# log(0.2) ≈ -1.6
# 그래서 손실은 앞에 마이너스를 붙여서 양수로 바꿔줍니다.
```

```python
import pandas as pd

fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head()  # fish에 있는 리스트 위에 5개 출력
```

```python
fish_input = fish[['Weight','Length','Diagonal','Height','Width']].to_numpy()
# Species(어종)을 제외한 나머지 필드를 fish_input 변수에 리스트 넣음
fish_target = fish['Species'].to_numpy()    # Species(어종) 정답파일로 생성
print(fish_target)
```

```python
from sklearn.model_selection import train_test_split

import numpy as np
np.set_printoptions(precision=6, suppress=True)   # 하단에 표준점수 변환용에서 사용함
#                   소수점 6자리까지 출력하겠다는 뜻입니다
#                                supppress=True : 과학적 표기법(지수표기) 사용 안함

# 사이킷런을 이용해서 훈련자료와 테스트자료를 나눔
train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42)
print('train_input')
print(train_input)    # 요소 119개, 특성 5
```

```python
print(train_input.shape)    # 요소 119개 특성 5
```

```python
print('test_input')
print(test_input)
```

```python
print(test_input.shape)    # 40개, 특성 5
```

```python
# 훈련용 119개, 특성 5, 테스트용 40개 특성 5

# 정답리스트를 보자.
print('train_input : ' + train_target)    # 훈련정답 119개 도출.
```

```python
print(test_target)    # 테스트 정답 40개
```

```python
from sklearn.preprocessing import StandardScaler
# 훈련세트와 테스트 세트의 특성을 표준화 전처리함 (표준점수로 처리)
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
print(train_scaled[:5])
```

```python
print(test_scaled[:5])
```

```python
from sklearn.linear_model import SGDClassifier # 확률적 경사 하강법 메서드로 이제 확인하자
# SGDClassifier : 사이킷런의 확률적 경사 하강법 제공, 2개의 매개변수를 지정
# loss='log_loss' : 로지스틱 손실 함수 지정
# max_iter=10 : 10회 에포크 반복 (전체 훈련세트를 10회 반복)
sc = SGDClassifier(loss='log_loss',max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)  # 훈련 시작, 학습 시작

print(sc.score(train_scaled, train_target)) # 77.31%
print(sc.score(test_scaled, test_target))   # 77.5%     -> 반복횟수 부족으로 max_iter 를 향상 후 훈련.
```

```python
# 확률적 경사 하강법은 점진적 학습이 가능하다라고 했음.
# SGDClssifier 객체를 또 만들지 말고 훈련할 모델을 sc로 추가하여 훈련을 추가한다.
sc.partial_fit(train_scaled, train_target)  # 부분적인 (partial) 학습

print(sc.score(train_scaled, train_target))  # 79.83%
print(sc.score(test_scaled, test_target))    # 77.5% -> 정확도가 향상되고 있음.
# 퍼센트가 증가 함 -> 무작정 부분적인 학습을 꼐속하는 것은 의미가 없다 -> 기준점을 제공해야 함.
# 에포크의 횟수가 기준점이 된다. 훈련 세트를 한 번 다 쓰는 것을 1 에포크라고 합니다.
# 현재는 등산 내려오다가 정지된 느낌
```

```python
# 에포크와 과대/과소적합
# 확률적 검사 하강법을 사용한 모델은 에포크 횟수에 따라서 과소, 과대 적합이 될 수 있다.
# 에포크 횟수가 적으면 훈련 세트를 덜 학습한다. (119개에서 10번만 랜덤추출하고 정지)
# 에포크 횟수가 충분히 많으면 훈련 세트를 완전히 학습할 것이다. (good)
```

```python
# 바꾸어 말하면 적은 에포크 횟수 동안에 훈련된 모델은 훈련세트와 테스트세트에 맞지 않음(과소적합)
# 반대로 에포크 횟수가 많은 훈련 모델인 경우에는 너무 잘 맞아서 테스트 세트에는 오히려 점수가 나쁨(과대적합)

# 과대 적합이 되기전에 훈련을 멈추는 것을 조기 종료함!!!
```

```python
import numpy as np

sc = SGDClassifier(loss='log_loss', random_state=42)

train_score = []    # 훈련세트 점수를 누적하는 빈 리스트 생성
test_score = []     # 테스트 세트 점수를 누적하는 빈 리스트 생성
# train_target에 있는 7개 생선목록을 제공
classes = np.unique(train_target)

# 파이썬의 _ 변수는 1회용
for _ in range(0, 300):   # 300번의 에포크 동안 훈련 진행
    sc.partial_fit(train_scaled, train_target, classes=classes)
    # 훈련을 부분적으로 추가하는 partial_fit

    # 반복할 때마다 훈련세트와 테스트 세트의 점수를 계산하여 train_score와 test_score 리스트에 추가함
    train_score.append(sc.score(train_scaled, train_target))     #훈련세트 점수를 빈 리스트에 추가
    test_score.append(sc.score(test_scaled, test_target))        #테스트세트 점수를 빈 리스트에 추가
```

```python
import matplotlib.pyplot as plt

plt.plot(train_score)   # 훈련세트 점수
plt.plot(test_score)    # 테스트세트 점수
plt.xlabel('epoch')     # 에포크 횟수
plt.ylabel('accuracy')  # 정확도
plt.show()

# 그래프 해석
# 그래프를 보면 보통 세 단계의 흐름이 나타납니다.
# 초반부 (Underfitting): 훈련 점수와 테스트 점수가 모두 낮고 같이 올라가는 단계. 아직 산을 더 내려가야 합니다.
# 중반부 (Best Fit): 두 선이 가장 높게 맞닿거나 간격이 좁은 지점. 우리가 멈춰야 할 에포크입니다.
# 후반부 (Overfitting): 훈련 점수는 계속 올라가는데, 테스트 점수가 떨어지기 시작하는 지점. 모델이 훈련 데이터의 잡음까지 외우기 시작했다는 신호입니다.
```

```python
# 결론 에포크 횟수를 100번 정도 돌리니 차트상으로 과대, 과소적합이 아님 (최상 good)
sc = SGDClassifier(loss='log_loss', max_iter=100, tol=None, random_state=42) # tol=None 조기종료 안함
# 확률적 검사 하강법은 일정 에포크 동안 성능이 향상되지 않으면 더 이상 훈련하지 않고 자동으로 조기종료.

sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))    # 95.79%
print(sc.score(test_scaled, test_target))      # 92.5% 로 최적의 상태로 분석됨
```

```python
sc.predict(test_scaled[:5])     # Perch Smelt Pike Perch Perch

proba = sc.predict_proba(test_scaled[:5])
np.round(proba, decimals=3)

# Bream', 'Parkki', 'Perch', 'Pike', 'Roach', 'Smelt', 'Whitefish']
# array([[0.   , 0.003, 0.774, 0.   , 0.219, 0.   , 0.005],
#        [0.   , 0.001, 0.015, 0.   , 0.04 , 0.945, 0.   ],
#        [0.   , 0.   , 0.011, 0.96 , 0.02 , 0.008, 0.001],
#        [0.   , 0.   , 0.525, 0.   , 0.423, 0.   , 0.052],
#        [0.   , 0.   , 0.688, 0.   , 0.306, 0.   , 0.006]])
```

```python
from sklearn.metrics import hinge_loss

train_score = []
test_score = []
train_loss = []
test_loss = []

classes = np.unique(train_target)
# partial_fit을 사용할 때 아주 중요한 역할을 합니다.
# 처음 등산을 시작할 때 "어디로 갈 수 있는지" 전체 지도를 미리 보여주는 것과 같습니다.
# partial_fit은 데이터를 한 번에 다 보지 않고 조금씩 나누어서 학습합니다.
# 여기서 문제가 발생할 수 있습니다.
# 모델의 정보 부족: 만약 첫 번째 미니 배치 데이터에 '도미(Bream)'와 '농어(Perch)'만 있고 '빙어(Smelt)'가 없다면,
# 모델은 세상에 빙어라는 생선이 존재한다는 사실을 모른 채 학습을 시작합니다.
# 에러 발생: 나중에 빙어 데이터가 들어오면 모델은 "어? 내 지도에는 이런 생선이 없는데?"라며 당황해서 에러를 냅니다.
# 해결책: 그래서 classes 매개변수에 전체 어종 리스트를 미리 알려주는 것

sc_hinge_test = SGDClassifier(loss='hinge', max_iter=1, tol=None, random_state=42)
# tol은 **Tolerance(허용 오차)**의 약자로, 쉽게 말해 **"언제 하산을 멈출지 결정하는 기준"**
# 경사 하강법으로 산을 내려가다 보면, 어느 순간 평지에 도달해서 아무리 걸어도 높이(Loss)가 거의 변하지 않는 지점이 옵니다. 이때 tol은 다음과 같이 동작합니다.
# 기본값 (예: 1e-3): "최근 몇 번의 학습 동안 손실(Loss)이 0.001보다 적게 줄어들면, 다 내려온 걸로 치고 학습을 조기 종료해라."라는 뜻
# 장점: 컴퓨터 자원을 아끼고, 과대적합이 되기 전에 알아서 멈춰줍니다.

# **"에포크(반복 횟수)에 따른 성능 변화를 관찰"**하는 것이 목적
# 강제 완주: tol=None으로 설정하면, 모델이 "오, 이제 다 내려온 것 같은데?"라고 판단하더라도 멈추지 않습니다.
# 이유: 우리가 설정한 max_iter 횟수(예: 300번)를 무조건 끝까지 다 실행하게 만들기 위해서입니다.
# 만약 tol을 설정해두면, 300번을 채우기도 전에 학습이 갑자기 멈춰버려서 그래프 뒷부분이 잘려 보일 수 있습니다.
# tol=None을 쓴 이유는, 모델이 제멋대로 학습을 중단하지 못하게 막고 우리가 정한 에포크 횟수만큼 그래프를 그리기 위해서

for _ in range(300):
    sc_hinge_test.partial_fit(train_scaled, train_target, classes=classes)

    # accuracy
    train_score.append(sc_hinge_test.score(train_scaled, train_target))
    test_score.append(sc_hinge_test.score(test_scaled, test_target))

    # loss
    train_decision = sc_hinge_test.decision_function(train_scaled)
    test_decision = sc_hinge_test.decision_function(test_scaled)

    train_loss.append(hinge_loss(train_target, train_decision))
    test_loss.append(hinge_loss(test_target, test_decision))

    # decision_function: 각 데이터가 결정 경계선에서 얼마나 떨어져 있는지 계산합니다. (Hinge Loss 계산을 위한 전 단계)
    # hinge_loss: "내가 정답을 맞혔더라도 아슬아슬하게 맞혔나, 아니면 확실하게 맞혔나?"를 수치화해서 저장합니다.
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))

# Accuracy (왼쪽 측)
ax1.plot(train_score,
         color='blue',
         linewidth=2.5,
         label='Train Accuracy')

ax1.plot(test_score,
         color='orange',
         linewidth=2.5,
         label='Train Accuracy')

ax1.set_xlabel('Epoch', fontsize=12)
ax1.set_ylabel('Accuracy',fontsize=12)
ax1.tick_params(axis='y')

# Loss(오른쪽 축)
ax2 = ax1.twinx()

ax2.plot(train_loss,
         color='green',
         linewidth=2,
         alpha=0.8,
         label='Train Loss')

ax2.plot(test_loss,
         color='red',
         linestyle='--',
         linewidth=2,
         alpha=0.8,
         label='Test Loss')

ax2.set_ylabel('Loss',fontsize=12)

# 범례 합치기
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='center right')

plt.title('Accuracy & Loss over Epochs', fontsize=14)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 투명도(alpha)를 조절해 격자를 은은하게 깔아주면, 데이터 선이 묻히지 않으면서도 수치를 정확히 읽을 수 있어 아주 세련된 시각화 방법


# 그래프 해석
# 가장 먼저 눈에 띄는 점은 **훈련 데이터(Blue)**와 **테스트 데이터(Orange)**의 정확도가 거의 겹쳐서 진행된다는 점
# 보통 훈련 점수가 훨씬 높고 테스트 점수가 낮으면 '과대적합(Overfitting)'을 의심하는데, 지금 그래프는 두 선이 매우 밀접하게 붙어 있습니다.
# 이는 모델이 데이터를 단순히 외우지 않고, 새로운 데이터에도 잘 적응하는 일반화 능력이 뛰어나다는 뜻입니다.
# 하산 완료 지점: 에포크 100 ~ 150 사이
# 정확도(Accuracy): 에포크 0에서 50까지 급격히 상승하다가, 100을 넘어서면서부터는 약 95% 근처에서 수렴(안정화)되는 모습입니다.
# 손실(Loss): 초반에 아주 높았던 손실값이 에포크 100 근처에서 바닥에 거의 붙었습니다. 이후 300까지는 아주 미세하게만 줄어듭니다.
# 결론: 이 데이터와 모델로는 에포크 100~150 정도면 충분히 학습이 완료되었다고 볼 수 있습니다. 300까지 돌리는 것은 약간의 자원 낭비일 수 있겠네요.
```

```python
# 조기종료 지점 찾아보기 (Early Stopping)
# test loss가 가장 낮은 epoch 또는 test loss가 patience 횟수만큼 연속 증가하면 멈춤
# minL(test) : test loss가 더 이상 감소하지 않는 시점을 찾습니다ㅏ.
```

```python
# 설정값
patience = 10
best_loss = float('inf')
patience_counter = 0
early_stop_epoch = 0

train_score_early = []
test_score_early = []
train_loss_early = []
test_loss_early = []

sc_hinge_test_early = SGDClassifier(loss='hinge',max_iter=1, tol=None, random_state=42)

for epoch in range(300):
   sc_hinge_test_early.partial_fit(train_scaled, train_target, classes=classes)

   # accuracy 저장
   train_score_early.append(sc_hinge_test_early.score(train_scaled, train_target))
   test_score_early.append(sc_hinge_test_early.score(test_scaled, test_target))

   # loss 계산
   train_decision = sc_hinge_test_early.decision_function(train_scaled)
   test_decision = sc_hinge_test_early.decision_function(test_scaled)

   current_test_loss = hinge_loss(test_target, test_decision)
   train_loss_early.append(hinge_loss(train_target, train_decision))
   test_loss_early.append(hinge_loss(test_target, test_decision))

   # Early Stopping 체크
   if current_test_loss < best_loss:
      best_loss = current_test_loss
      patience_counter = 0
      early_stop_epoch = epoch
   else:
      patience_counter += 1

   if patience_counter >= patience:
      print(f"Early stopping at epoch {epoch}")
      break
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))

# Accuracy
ax1.plot(train_score_early, color='blue', linewidth=2, label='Train Acc')
ax1.plot(test_score_early, color='orange', linewidth=2, label='Test Acc')

# Early stopping 세로선
ax1.axvline(x=early_stop_epoch,
            color='black',
            linestyle=':',
            linewidth=2,
            label='Early Stop')

ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')

# Loss
ax2 = ax1.twinx()
ax2.plot(train_loss_early, color='green', linestyle='--', label='Train Loss')
ax2.plot(test_loss_early, color='red', linestyle='--', label='Test Loss')
ax2.set_ylabel('Loss')

# 범례 통합
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='center right')

plt.title('Accuracy & Loss with Early Stopping')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
