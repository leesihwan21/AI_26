```python
import os
import shutil
from tqdm import tqdm

# 1. 경로 설정 (확인된 폴더 이름 사용)
base_dir = "/content/pothole"
new_dir = "/content/pothole_v2"

# 2. 병합 실행
for split in ['train', 'val', 'test']:
    for folder in ['images', 'labels']:
        src = os.path.join(new_dir, split, folder)
        dst = os.path.join(base_dir, split, folder)

        # 목적지 폴더가 없으면 생성
        os.makedirs(dst, exist_ok=True)

        if os.path.exists(src):
            print(f"[{split}/{folder}] 데이터를 통합 중입니다...")
            files = os.listdir(src)
            for f in tqdm(files):
                # 파일명 중복 방지를 위해 v2_ 접두어를 붙여 복사합니다.
                shutil.copy(os.path.join(src, f), os.path.join(dst, f"v2_{f}"))

print("\n✨ 모든 병합이 완료되었습니다!")

# 3. 최종 숫자 확인 (0 탈출 체크!)
print("--- 최종 통합 학습 이미지 개수 ---")
!ls -l /content/pothole/train/images | grep ^- | wc -l
```

```python
# 1. 필수 라이브러리 설치
!pip install ultralytics

import os
import yaml
from ultralytics import YOLO

# 2. 데이터가 있는 'pothole_v2' 폴더를 기준으로 경로 설정
root_dir = '/content/pothole_v2'  # 데이터가 있는 폴더로 변경
yaml_path = os.path.join(root_dir, 'data.yaml')

# 설계도(YAML)가 pothole_v2 안에 있는지 확인하고 경로 수정
if os.path.exists(yaml_path):
    with open(yaml_path, 'r') as f:
        config = yaml.safe_load(f)

    # 실제 파일이 들어있는 경로로 강제 지정
    config['train'] = os.path.join(root_dir, 'train/images')
    config['val'] = os.path.join(root_dir, 'valid/images')  # 'valid' 폴더명 반영
    config['test'] = os.path.join(root_dir, 'test/images')

    with open(yaml_path, 'w') as f:
        yaml.dump(config, f)
    print("✅ pothole_v2 기준 경로 최적화 완료!")
else:
    # 만약 pothole_v2 안에 yaml이 없다면 pothole의 것을 빌려옵니다.
    print("ℹ️ pothole_v2 내 yaml이 없어 설정을 새로 생성합니다.")
    config = {
        'train': os.path.join(root_dir, 'train/images'),
        'val': os.path.join(root_dir, 'valid/images'),
        'test': os.path.join(root_dir, 'test/images'),
        'nc': 1,
        'names': ['pothole']
    }
    with open(yaml_path, 'w') as f:
        yaml.dump(config, f)

# 3. 학습 시작
model = YOLO('yolov8n.pt')

model.train(
    data=yaml_path,
    epochs=50,
    imgsz=160,
    device='cpu',
    project='ITS-Guard-Final',
    name='Pothole_V2_Success'
)


# 결과 해석
# 정밀도 (Precision) : 0.816(81.6%)
# 모델이 포트홀이라고 판단한 것 중 실제 포트홀 일 확률이 매우 높다. 오탐지가 적다는 뜻으로 운전자에게
# 잘못된 알림을 줄 확률이 낮아 긍정적
# 재현율 (Recall) : 0.420(42.0%)
# 실제 존재하는 포트홀 중 약 42% 정도를 찾아내고 있다. 현재 CPU 환경과 작은 이미지 사이즈(Imsz=160)의 한계로 보이며,
# 중요한 포트홀을 놓치지 않으려면 향후 데이터 증강이나 해상도 상향이 필요할 수 있다.
# mAP50: 0.526(52.6%)
# 포트홀 탐지의 정확도를 나타내는 종합 지표.처음 시작하는 프로토타입 단계치고는 준수한 결과이다.

# 이 데이터를 기반으로 해상도 상향이 필요한 이유
# 현재 이 데이터 이미지 사이즈는 가로세로 160픽셀로 아주 작게 줄여서 본다는 뜻이다.
# 문제점 : 도로 사진에서 멀리 있는 작은 포트홀은 픽셀 몇 개 수준으로 뭉개져 버린다. 모델 입장에서는 이게 포트홀인지 그냥 그림자인지 구분하기 어렵다.
# 따라서, 해결책 : 보통 고성능 모델에서는 imgsz=640 을 표준으로 사용한다. 하지만 해상도를 높이면 미세한 균열이나 작은 구멍도 선명하게 포착되어
# 재현율이 비약적으로 상승한다.

# 데이터 증강 및 추가 데이터가 필요한 이유
# 데이터셋 지표를 보면 클래스당 개체 수가 약 240개 정도이다. 딥러닝 모델이 다양한 환경에 적응하기에는 다소 적은 양이라 볼 수 있다.
# 필요한 데이터 이미지 : 비오는 날 밤, 그림자가 짙게 깔린 도로, 각도가 비스듬한 도로등 다양한 환경의 이미지가 추가되면 좋을 것 같다.
# 증강 기법 : 기존 이미지를 좌우 반전하거나, 밝기를 조절하거나, 약간의 노이즈를 섞는 증강을 통해 모델이 이런 상황도 포트홀이구나 라고 학습하게 만들어야한다.

# 정확도를 조금 더 올리기 위한 Set up 전략.
# 데이터 증강 기법과 해상도 상향은 정확도 측면에서 질을 완전히 바꿔놓을 수 있는 기술이다.
# 해상도(imgsz) 160 -> 640 : 현재는 포트홀을 멀리서 본 흐릿한 점으로 인식한다면, 640으로 높이면 표면의 질감과 깊이까지 보게 된다.
# 데이터 증강(Augmentation) : 현재는 깨끗한 사진의 위주라면, 여기에 노이즈 이미지 데이터를 섞거나 일부러 Blur(흐릿) 하게 만든 사진을 학습시키면 모델이 훨씬 더 똑똑해진다.

# results_dict를 기준으로 분석 결과
# Precision(정밀도) 81.6% 모델이 "포트홀이다!"라고 했을 때, 실제로 포트홀일 확률입니다. 매우 높은 편이며, 엉뚱한 걸 보고 알람을 울릴 일이 적다는 뜻입니다.
# Recall (재현율)	42.0% 실제 길에 있는 포트홀 10개 중 모델이 찾아낸 비율입니다. 현재 모델은 4개 정도를 찾아내고 6개는 놓치고 있습니다. (해상도 영향이 큼)
# mAP50	52.6% 포트홀을 얼마나 정확한 위치에 박스를 치는지 보여주는 종합 점수입니다. 입문 단계에서 50%를 넘긴 것은 꽤 성공적인 학습입니다.
# Inference (추론 속도)	20.2ms 사진 한 장을 분석하는 데 걸리는 시간입니다. 약 0.02초로, 실시간 영상 주행 시에도 무리 없이 작동할 만큼 아주 빠릅니다
```
